## **TEORIA**

### **ZASADY GRY W BASEBALL**

**Zespoły**: Grają dwie drużyny, z których jedna działa jako drużyna atakująca (batsmen) i próbuje zdobyć punkty, a druga działa jako drużyna broniąca (pitchers i fielders), starając się uniemożliwić zdobycie punktów przeciwnikowi.

**Rzuty**: Gra zaczyna się od rzutu (pitch) ze strony miotacza (pitcher) w kierunku pałkarza (batter) z drużyny atakującej, który stoi przy bazie.

**Bazy**: Na boisku znajdują się cztery bazy ustawione w kształt diamentu: pierwsza, druga, trzecia baza oraz baza domowa. Celem pałkarza jest przebiegnięcie wszystkich baz i dotarcie do bazy domowej, zdobywając tym samym punkt.

**Uderzenie**: Pałkarz próbuje uderzyć piłkę baseballową przy użyciu kija po rzucie miotacza. Jeśli uda mu się uderzyć, zaczyna biec w kierunku pierwszej bazy.

**Fielding**: Drużyna broniąca stara się przechwycić uderzoną piłkę i jak najszybciej przekazać ją między bazami w celu wyeliminowania pałkarza lub innych zawodników atakujących.

**Inningi**: Gra jest podzielona na dziewięć inningów, z których każdy składa się z kolejki ataku i obrony obu drużyn. Drużyna atakująca ma ograniczoną ilość prób (autsów) na zdobycie punktu, zanim nastąpi zamiana stron.

**Punkty**: Zdobywa się punkty, gdy pałkarz przebiegnie wszystkie cztery bazy i dotknie bazy domowej. Drużyna z większą liczbą zdobytych punktów wygrywa.

**Outy**: Pałkarz może być wyeliminowany z gry poprzez trzy błędy (outs), na przykład przez złapanie piłki przez obrońców zanim dotknie ziemi, rzut do bazy przed pałkarzem lub trzy niemalowane rzuty (strikes). [1]

### **RODZAJE RZUTÓW BASEBALLOWYCH**

W baseballu wyróżniamy rózne techniki rzutów, w którym piłka przyjmuje swoją charakterystyczną trajektorię. Różne rodzaje rzutów są stosowane przez miotaczy w celu zmylenia przeciwnika i utrudnienia trafienia piłki przez pałkarza. Rzuty baseballowe mogą przyjmować rózne formy, z róznymi stanami początkowymi, a my przedstawimy najpopularniejsze z nich.

**Fastball** to najczęściej stosowany typ rzutów. Piłka wyrzucona jest z możliwie jak największą prędkością i podąża po torze prostoliniowym w kierunku pałkarza. [2]

**Slider** to rzut, w którym piłka który porusza się bocznie i w dół do strefy uderzeniowej miotacza. [3]

**Screwball** to rzut, w któym piłka łamie w przeciwnym kierunku niż slider. [4]

**Curveball** to rzut, który nadaje piłce ruch obrotowy, powodując, że nurkuje ona w kierunku pałkarza. [5]

### **ZAŁOŻENIA I OPIS RÓWNAŃ**

Analizując trajektorię lotu piłki baseballowej będzie nam potrzebna całkowita siła działająca na piłkę podczas lotu od miotacza do pałkarza. Całkowitą siłę piłki możemy rozłożyć na siłe grawitacji, siłe oporu powietrza i siłe Magnusa.

Siłę grawitacji możemy zapisać jako:
<br>
$$F_g = -mgẑ$$ 



Gdzie:
* $m$ - masa piłki
* $g$ - przyspieszenie ziemskie
* $ẑ$ - wektor $[0,0,1]$
  

***

Siłę oporu powietrza na piłkę możemy zapisać jako:
<br>
$$F_d = mv\mathbf{v}f(v)$$

Gdzie:
* $m$ - masa piłki
* $v$ - prędkość piłki 
* $\mathbf{v}$ - wektor $[v_x,v_y,v_z]$


* $f(v)$ - funkcja zadana wzorem:

$$f(v) = 0.0039 + \frac{0.0058}{1 + e^{\left(\frac{v - v_d}{\Delta}\right)}}$$

Gdzie:
* $v_d = 35$ $m/s$
* $\Delta = 5$ $m/s$ 

***

Siłę Magnusa możemy zapisać jako:
$$F_m = S(v)\bm{\omega} \times \mathbf{v}$$

Gdzie:
* $S(v) = Bm = (4.1*10^{-4})m$ - sugerowane przybliżenie dla typowego rzutu baseballowego [6]
* $\bm{\omega}$ - wektor prędkości kątowej $\omega[0,\sin{\alpha},\cos{\alpha}]$ ( $\omega$ - prędkość kątowa )
* $\mathbf{v}$ - wektor $[v_x,v_y,v_z]$
* $\times$ - iloczyn wektorowy


## **ROZWIĄZANIE ANALITYCZNE**

Z drugiej zasady dynamiki Newtona wiemy, że całkowitą siłę piłki możemy zapisać jako siłę wypadkową trzech sił: siły grawitacji, siły oporu ruchu, siły Magnusa

$$F_w = F_g + F_d + F_m$$

$$am = -mgẑ -  mv\mathbf{v}f(v) + S(v)\bm{\omega} \times \mathbf{v}$$

$$a = -gẑ -  v\mathbf{v}f(v) + \frac{S(v)}{m}\bm{\omega} \times \mathbf{v}$$

Wiemy, że: $a = \frac{dv}{dt}$

$$\frac{dv}{dt} = -g\hat{z} -  v\mathbf{v}f(v) + \frac{S(v)}{m}\bm{\omega} \times \mathbf{v}$$

Wiedząc,że:
* $\bm{\omega}$ - wektor prędkości kątowej $\omega[0,\sin{\alpha},\cos{\alpha}]$
* $\mathbf{v}$ - wektor $[v_x,v_y,v_z]$


Możemy zapisać iloczyn wektorory $\bm{\omega} \times \mathbf{v}$ jako:


$$\bm{\omega} \times \mathbf{v} = \omega(\sin{\alpha}v_z - \cos{\alpha}v_y)\hat{x} + \omega\cos{\alpha}v_x\hat{y} - \omega\sin{\alpha}v_x\hat{z}$$

Zapiszmy przyspieszenie $a$ jako składowe $a_x = \frac{dv_x}{dt}$, $a_y = \frac{dv_y}{dt}$, $a_z =\frac{dv_z}{dt}$ :

Czyli mamy:
<br>
* dla $a_x$ :
$$\frac{dv_x}{dt} = - vv_xf(v) + \frac{S(v)}{m}\omega(\sin{\alpha}v_z - \cos{\alpha}v_y)$$

* dla $a_y$ :
$$\frac{dv_y}{dt} = - vv_yf(v) + \frac{S(v)}{m}\omega\cos{\alpha}v_x$$

* dla $a_z$ :
$$\frac{dv_z}{dt} = -g -  vv_zf(v) - \frac{S(v)}{m}\omega\sin{\alpha}v_x$$

***

Ostatecznie równania ruchu piłki baseballowej rzuconej przez miotacz można opisać za pomocą następujących
sześciu równań różniczkowych zwyczajnych:

$$\frac{dx}{dt} = v_x$$

$$\frac{dy}{dt} = v_y$$

$$\frac{dz}{dt} = v_z$$

$$\frac{dv_x}{dt} = - vv_xf(v) + \frac{S(v)}{m}\omega(\sin{\alpha}v_z - \cos{\alpha}v_y)$$

$$\frac{dv_y}{dt} = - vv_yf(v) + \frac{S(v)}{m}\omega\cos{\alpha}v_x$$

$$\frac{dv_z}{dt} = -g -  vv_zf(v) - \frac{S(v)}{m}\omega\sin{\alpha}v_x$$

Dla warunków brzegowych:

* $x(t=0) = 0$
* $y(t=0) = 0$
* $z(t=0) = 0$
* $v_x(t=0) = v_0\cos{\theta}$
* $v_y(t=0) = 0$
* $v_z(t=0) = v_0\sin{\theta}$

***

Szczegółową analizę równań różniczkowych przeprowadzimy na *fastballu*.

Prędkość piłki podczas wykonywania *fastballu* zazwyczaj sięga od 153 do 169 km/h. [2]

W naszym modelu weźmiemy średnią prędkość, czyli:
$$v_ś = \frac{153+169}{2} = 161 \frac{km}{h}$$ 

Dla $v_ś = 161 \frac{km}{h}= \frac{161 \cdot 1000}{3600} \frac{m}{s} \approx 44.7222 \frac{m}{s} $

$$f(v_ś) = f(44.7222)= 0.0039 + \frac{0.0058}{1 + e^{\left(\frac{44.7222 - 35}{5}\right)}} \approx 0.004625$$

$ \omega = 1800 \frac{obr.}{min}  = 1800 \cdot\frac{2\pi}{60} \frac{rad}{s} \approx 188.496 \frac{rad}{s}$

Zapiszmy $a_x = \frac{dv_x}{dt}$, $a_y = \frac{dv_y}{dt}$, $a_z =\frac{dv_z}{dt}$ podstawiając $v_ś \approx 44.7222 \frac{m}{s}$,  $ f(v_ś) \approx 0.004625$ oraz $ \omega \approx 188.496 \frac{rad}{s}$

* $a_x = \frac{dv_x}{dt}$

$$\frac{dv_x}{dt} = - 44.7222v_x0.004625 + 4.1\cdot10^{-4}\cdot188.496(\sin{\alpha}v_z - \cos{\alpha}v_y)$$

dla $\alpha = 225\degree$:


$$\frac{dv_x}{dt} = - 0.2068v_x + 0.00041\cdot188.496(\sin{(225\degree)}v_z - \cos{(225\degree)}v_y)$$

$\sin{(225\degree)} = -0.707$, $\cos{(225\degree)} = -0.707$

$$\frac{dv_x}{dt} = - 0.2068v_x + 0.07728336(-0.707v_z + 0.707v_y)$$

$$\frac{dv_x}{dt} = - 0.2068v_x - 0.0546393v_z + 0.0546393v_y$$

$$\frac{dv_x}{dt} = - 0.2068v_x  + 0.0546393v_y - 0.0546393v_z$$

Podobnie robimy dla $a_y = \frac{dv_y}{dt}$, $a_z =\frac{dv_z}{dt}$ i otrzymujemy odpowiednio

* $a_y = \frac{dv_y}{dt}$

$$\frac{dv_y}{dt} = - 0.0546393v_x - 0.2068v_y$$

* $a_z = \frac{dv_z}{dt}$ (tutaj dodatkowo zakładamy, że $g=9.81\frac{m}{s^2}$)


$$\frac{dv_z}{dt} = 0.0546393v_x - 0.2068v_z - 9.81$$

Ostatecznie mamy układ równań:


$$\frac{dv_x}{dt} = - 0.2068v_x  + 0.0546393v_y - 0.0546393v_z$$

$$\frac{dv_y}{dt} = - 0.0546393v_x - 0.2068v_y$$

$$\frac{dv_z}{dt} = 0.0546393v_x - 0.2068v_z - 9.81$$

Rozwiązując ten układ równań różniczkowych dostajemy rozwiązanie analityczne:

$$\left\{\begin{gathered}v_{x}=\dfrac{\sqrt{2}\,C\,\cos\left(\frac{546393\,t}{78125\cdot 64\,\sqrt{2}}\right)-\sqrt{2}\,C_{1}\,\sin\left(\frac{546393\,t}{78125\cdot 64\,\sqrt{2}}\right)}{{e}^{\frac{517\,t}{2500}}}+\dfrac{26800576650000}{2436857310449}\\v_{y}=\dfrac{-C\,\sin\left(\frac{546393\,t}{78125\cdot 64\,\sqrt{2}}\right)-C_{1}\,\cos\left(\frac{546393\,t}{78125\cdot 64\,\sqrt{2}}\right)+C}{{e}^{\frac{517\,t}{2500}}}-\dfrac{7321823738761725}{2519710459004266}\\v_{z}=\dfrac{C\,\sin\left(\frac{546393\,t}{78125\cdot 64\,\sqrt{2}}\right)+C_{1}\,\cos\left(\frac{546393\,t}{78125\cdot 64\,\sqrt{2}}\right)+C}{{e}^{\frac{517\,t}{2500}}}-\dfrac{112206027338761725}{2519710459004266}\end{gathered}\right.$$

W celu ułatwienia naszych obliczeń założymy, że znajdujemy się w przestrzeni dwuwymiarowej.

Znając prękość początkową i kąt, pod którym wyrzucamy piłkę obliczymy $v_x$ i $x_z$ w czasie $t=0$

Wiemy, że

$$v_x=v_0cos(\theta)$$

$$v_y=0$$

$$v_z=v_0sin(\theta)$$

Załóżmy, że kąt $\theta$ ma wartość $1\degree$.

Zakładamy, że kąt jest bardzo mały, ponieważ im kąt wyrzutu piłki jest mniejszy, tym piłka jest cięższa do przejęcia.

$v_0 = 95 \frac{mil}{h} \approx 152.855 \frac{km}{h} =  \frac{152.855 \cdot 1000}{3600} \frac{m}{s} \approx 42.459 \frac{m}{s} $

Podstawiając nasze wiadome wartości mamy:

$$v_x = 42.459\cdot \cos(1\degree)$$

$$v_x \approx 42.453\frac{m}{s}$$

$$v_z = 42.459\cdot\sin(1\degree)$$

$$v_z \approx 0.741\frac{m}{s}$$

*** 


Wiedząc że:

$$v_x(0) = 42.453\frac{m}{s}$$

$$v_z(0) = 0.741\frac{m}{s}$$

Oraz, że:

$$\frac{dv_x}{dt} = - 0.2068v_x - 0.0546393v_z$$

$$\frac{dv_z}{dt} = 0.0546393v_x - 0.2068v_z - 9.81$$

Otrzymujemy rozwiązanie metodą analityczną: [7]

$$\left\{\begin{gathered}v_{x}(t)=-\dfrac{403220735\,\sin\left(\frac{546393\,t}{10000000}\right)}{8944023\,{e}^{\frac{517\,t}{2500}}}+\dfrac{88700263\,\cos\left(\frac{546393\,t}{10000000}\right)}{2885750\,{e}^{\frac{517\,t}{2500}}}+\dfrac{53601153300000}{4575169310449}\\v_{z(t)}=\dfrac{88700263\,\sin\left(\frac{546393\,t}{10000000}\right)}{2885750\,{e}^{\frac{517\,t}{2500}}}+\dfrac{403220735\,\cos\left(\frac{546393\,t}{10000000}\right)}{8944023\,{e}^{\frac{517\,t}{2500}}}-\dfrac{202870800000000}{4575169310449}\end{gathered}\right.$$

Używając tych równań jesteśmy w stanie napisać program przedstawiający trajektorię wyrzuconej piłki.

Pisząc nasze programy założymy, że piłka wyrzucana jest na wysokości $1.1 m$ nad ziemią, a oś $X$ jest ograniczona po prawej stronie przez $18.44m$, ponieważ jest to przepisowa odległość miotacza od pałkarza w baseballu. [1]

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Funkcje opisujące równania ruchu
def vx(t):
    return (-403220735 * np.sin(546393 * t / 10000000) / 8944023 / np.exp(517 * t / 2500) +
            88700263 * np.cos(546393 * t / 10000000) / (2885750 * np.exp(517 * t / 2500)) +
            53601153300000 / 4575169310449)

def vz(t):
    return (88700263 * np.sin(546393 * t / 10000000) / (2885750 * np.exp(517 * t / 2500)) +
            403220735 * np.cos(546393 * t / 10000000) / 8944023 / np.exp(517 * t / 2500) -
            202870800000000 / 4575169310449)

# Parametry symulacji
t_values = np.linspace(0, 10, 1000)  # Zakres czasu od 0 do 10 sekundy z 1000 krokami

# Obliczenia trajektorii ruchu
x_values = np.cumsum(vx(t_values)) * (t_values[1] - t_values[0])
z_values = np.cumsum(vz(t_values)) * (t_values[1] - t_values[0])

# Rysowanie trajektorii
plt.plot(x_values, z_values)
plt.title('Trajektoria lotu fastballa w 2D (metoda analityczna)')
plt.xlabel('Odległość miotacza od pałkarza (m)')
plt.ylabel('Wysokość (m)')
plt.xlim([0, 18.44])  
plt.ylim([-3, 1])    
plt.grid(True)
plt.show()
```

# **ROZWIĄZANIA METODAMI NUMERYCZNYMI**

Oprócz podejścia analitycznego do rozwiązywania równań różniczkowych, istnieją również metody numeryczne, które pozwalają na przybliżone znalezienie rozwiązania. W tej części omówimy Metodę Eulera i Rungego-Kutty'ego oraz porównamy je z naszym rozwiązaniem analitycznym.

## **Metoda Eulera**

Metoda Eulera stanowi podstawową procedurę numeryczną pierwszego rzędu, wykorzystywaną do rozwiązywania równań różniczkowych zwyczajnych z zadaną wartością początkową. Wartość kolejnego punktu obliczeniowego jest uzależniona od poprzednich. W kontekście Metody Eulera istnieją dwie wersje: standardowa i zmodyfikowana. W tym przypadku skoncentrujemy się na metodzie standardowej.



Metoda ta jest opisana wzorem rekurencyjnym: [8]


$$ y_{n+1} = y_{n}+hf(x_{n},y_{n}).$$



Poniżej przedstawiamy rozwiązanie naszego układu równań różniczkowych dla *fastballa*, w którym zastosowałyśy metodę Eulera z krokiem czasowym $h=0.01$. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry
v_s = 44.7222 
v0 = 42.459
theta = 1
omega = 188.496
alpha = 225
g = 9.81

# Funkcja opisująca f(v)
def f(v):
    return 0.0039 + 0.0058 / (1 + np.exp((v - 35) / 5))

# Układ równań różniczkowych
def equations(vx, vy, vz, v_s):
    
    dvx_dt = -v_s * vx * f(v_s) + 4.1 * 10**(-4) * omega * (np.sin(np.radians(alpha)) * vz - np.cos(np.radians(alpha)) * vy)
    dvy_dt = -v_s * vy * f(v_s) + 4.1 * 10**(-4) * omega * np.cos(np.radians(alpha)) * vx
    dvz_dt = -g - v_s * vz * f(v_s) - 4.1 * 10**(-4) * omega * np.sin(np.radians(alpha)) * vx
    return dvx_dt, dvy_dt, dvz_dt

# Metoda Eulera
def euler_method(v0, theta, omega, alpha, g, h, steps, v_s):
    vx, vy, vz = v0 * np.cos(np.radians(theta)), 0, v0 * np.sin(np.radians(theta))
    x, y, z = [0], [0], [0]
    
    for _ in range(steps):
        dvx_dt, dvy_dt, dvz_dt = equations(vx, vy, vz, v_s)
        vx += h * dvx_dt
        vy += h * dvy_dt
        vz += h * dvz_dt
        x_new = x[-1] + h * vx
        if x_new > 19.44:
            break  # Przerwij pętlę, jeżeli x przekroczy 19.44
        x.append(x_new)
        y.append(y[-1] + h * vy)
        z.append(z[-1] + h * vz)
    
    return x, y, z

# Parametry symulacji
h = 0.01  # krok czasowy
steps = 1000

# Symulacja przy użyciu metody Eulera
x_euler, y_euler, z_euler = euler_method(v0, theta, omega, alpha, g, h, steps, v_s)

plt.plot(x_euler, z_euler)
plt.xlabel('Odległość miotacza od pałkarza (m)')
plt.ylabel('Wysokość (m)')
plt.ylim(-3,1)
plt.xlim(0,18.44)
plt.title('Trajektoria lotu fastballa w 2D (Metoda Eulera)')
plt.grid(True)
plt.show()
```

## **Metoda Rungego-Kutty**

Metoda Rungego-Kutty czwartego rzędu jest jedną z metod numerycznych stosowanych do rozwiązywania równań różniczkowych zwyczajnych. Jej geneza wiąże się z rozwinięciem w szereg Taylora, co pozwala na osiągnięcie znaczącej precyzji przy  umiarkowanej złożoności obliczeniowej.

$$y_{i+1} = y_{i}+\frac{1}{6}(k_{1}+k_{2}+k_{3}+k_{4})h$$
	
gdzie,
	
$$k_{1}=f(x_{i},y_{i})$$
$$k_{2}= f(x_{i} + \frac{1}{2}h,y_{i} + \frac{1}{2}k_{1}h)$$
$$k_{3}=f(x_{i} +  \frac{1}{2}h,y_{i} + \frac{1}{2}k_{2}h)$$
$$k_{4}=f(x_{i}+h,y_{i} +k_{3}h)$$

Poniżej przedstawiamy rozwiązanie naszego układu równań różniczkowych dla *fastballa*, w którym zastosowałyśy metodę Rungego-Kutty. 

```python
import numpy as np
import matplotlib.pyplot as plt


# Funkcja opisująca f(v)
def f(v):
    return 0.0039 + 0.0058 / (1 + np.exp((v - 35) / 5))

# Układ równań różniczkowych
def equations(vx, vy, vz, v_s, omega, alpha, g):
    
    dvx_dt = -v_s * vx * f(v_s) + 4.1 * 10**(-4) * omega * (np.sin(np.radians(alpha)) * vz - np.cos(np.radians(alpha)) * vy)
    dvy_dt = -v_s * vy * f(v_s) + 4.1 * 10**(-4) * omega * np.cos(np.radians(alpha)) * vx
    dvz_dt = -g - v_s * vz * f(v_s) - 4.1 * 10**(-4) * omega * np.sin(np.radians(alpha)) * vx
    return dvx_dt, dvy_dt, dvz_dt

# Metoda Rungego-Kutty czwartego rzędu 
def runge_kutta_4th_order(v0, theta, omega, alpha, g, h, steps, v_s):
    vx, vy, vz = v0 * np.cos(np.radians(theta)), 0, v0 * np.sin(np.radians(theta))
    x, y, z = [0], [0], [0]
    
    for _ in range(steps):
        k1x, k1y, k1z = h * np.array(equations(vx, vy, vz, v_s, omega, alpha, g))
        k2x, k2y, k2z = h * np.array(equations(vx + k1x/2, vy + k1y/2, vz + k1z/2,v_s, omega, alpha, g))
        k3x, k3y, k3z = h * np.array(equations(vx + k2x/2, vy + k2y/2, vz + k2z/2, v_s, omega, alpha, g))
        k4x, k4y, k4z = h * np.array(equations(vx + k3x, vy + k3y, vz + k3z, v_s, omega, alpha, g))
        
        vx += (k1x + 2*k2x + 2*k3x + k4x) / 6
        vy += (k1y + 2*k2y + 2*k3y + k4y) / 6
        vz += (k1z + 2*k2z + 2*k3z + k4z) / 6
        x_new = x[-1] + h * vx
        if x_new > 19.44:
            break  # Przerwij pętlę, jeżeli x przekroczy 19.44
        x.append(x_new)
        
        y.append(y[-1] + h * vy)
        z.append(z[-1] + h * vz)
    
    return x, y, z
# Symulacja przy użyciu metody Rungego-Kutty czwartego rzędu
x_rk4, y_rk4, z_rk4 = runge_kutta_4th_order(v0 =42.459, theta = 1, omega = 188.496, alpha =225, g=9.81, h =0.03, steps =1000, v_s =44.7222)

plt.plot(x_rk4, z_rk4)
plt.xlabel('Odległość miotacza od pałkarza (m)')
plt.ylabel('Wysokość (m)')
plt.ylim(-3,1)
plt.xlim(0,18.44)
plt.title('Trajektoria lotu fastballa w 2D (Metoda Rungego-Kutty 4. rzędu)')
plt.grid(True)
plt.show()
```

### **KNUCKLE BALL**

Ciekawym rodzajem rzutu jest *knuckle ball*, czyli rzucenie piłki z niewielką, w porównaniu do innych rzutów, prędkośćią (około 65 mil na godzinę) i możliwie jak najmniejszą rotacją w czasie lotu. Rzucenie piłki baseballowej w taki sposób sprawia, że jej trajektoria jest nieregularna i nieprzewidywalna, dzięki czemu jest ona bardzo trudna do wyłapania przez pałkarza. [10]

Ta nieregularność bierze się z siły bocznej dzialającej na piłkę. Siła boczna natomiast bierze się z różnicy w działaniu siły opru na piłkę w zależnośći od tego czy strona jest gładka, czy przechodzi przez nią szew. Siła oporu po gładkiej stronie piłki jest większa niż po stronie ze szwem, a działająca siła boczna skierowana jest w stronę szwu. Gdy piłka leci w stronę pałkarza, zmienia się jej orientacja, strona ze szwem przesuwa się z boku na bok, a wraz z nią siła boczna. Jeśli piłka obraca się szybko oscylacje się niwelują. Jednak w przypadku *knuckle balla* są one znaczące. [11]

Aby przedstawić nieprzewidywalne trajektorie *knuckle balla* wykorzystamy wzór na siłę boczną wyznaczoną przez Wattsa i Sawyera dla piłki poruszającej się z prędkościa $65\frac{mil}{s}$:

$$G(\varphi)=0.5[sin(4\varphi) - 0.25sin(8\varphi) + 0.08sin(12\varphi) - 0.025sin(16\varphi)]$$

Gdzie $\varphi$ to położenie kątowe piłki.

Uwzględniając siłę boczną i to, że przy bardzo małej rotacji siła Magnusa jest pomijalnie mała, możemy wyznaczyć równania rózniczkowe dla *knuckle balla*:

$$\frac{dx}{dt} = v_x$$

$$\frac{dy}{dt} = v_y$$

$$\frac{dz}{dt} = v_z$$

$$\frac{dv_x}{dt} = -f(v)vv_x$$

$$\frac{dv_y}{dt} = -f(v)vv_y + gG(\varphi)$$

$$\frac{dv_z}{dt} = -f(v)vv_z - g$$

$$\frac{d\varphi}{dt} = \omega

Dla warunków początkowych:

* $x(t=0) = 0$
* $y(t=0) = 0$
* $z(t=0) = 0$
* $v_x(t=0) = v_0\cos{\theta}$
* $v_y(t=0) = 0$
* $v_z(t=0) = v_0\sin{\theta}$
* $\varphi(t=0) = \varphi_0$ 

Aby przedstawić jak początkowe położenie kątowe piłki wpływa na trajektorię ruchu *knuckle balla* napiszemy program wykorzystując metodę Runge-Kutty. Ustalmy wartości dla każdej próby:

$v_0= 65\frac{mil}{s} \approx 29\frac{m}{s}$, 

$\theta=4\degree$, $g = 9.81 \frac{m}{s^2}$

$\omega = 20\frac{obr}{min} \approx 2\frac{rad}{s}$, 

$v_ś = \frac{97 + 113}{2}\frac{km}{h} \approx 29 \frac{m}{s}$ [10]

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection


v0 = 29.0
theta = 4.0  
omega = 2.0
g = 9.81 
v = 29

# Różne wartości początkowe dla phi
phi_values = [0, 18, 36, 54, 72]

def f(v):
    return 0.0039 + 0.0058 / (1 + np.exp((v - 35) / 5))

# Funkcja opisująca G(phi)
def G(phi):
    return 0.5 * (np.sin(4 * phi) - 0.25 * np.sin(8 * phi) + 0.08 * np.sin(12 * phi) - 0.025 * np.sin(16 * phi))

# Układ równań różniczkowych
def equations(v, phi):
    dvx_dt = -f(v) * v * np.cos(phi)
    dvy_dt = -f(v) * v * np.sin(phi) + g * G(phi)
    dvz_dt = -f(v) * v * np.sin(phi) - g
    dphi_dt = omega

    return dvx_dt, dvy_dt, dvz_dt, dphi_dt

# Metoda Rungego-Kutty czwartego rzędu (RK4)
def runge_kutta_4th_order(v0, theta, omega, phi0, g, h, steps):
    vx, vy, vz, phi = v0 * np.cos(np.radians(theta)), 0, v0 * np.sin(np.radians(theta)), phi0
    x, y, z, phi_list = [0], [0], [0], [phi]
    
    for _ in range(steps):
        k1x, k1y, k1z, k1phi = h * np.array(equations(v0, phi))
        k2x, k2y, k2z, k2phi = h * np.array(equations(v0, phi + k1phi/2))
        k3x, k3y, k3z, k3phi = h * np.array(equations(v0, phi + k2phi/2))
        k4x, k4y, k4z, k4phi = h * np.array(equations(v0, phi + k3phi))
        
        vx += (k1x + 2*k2x + 2*k3x + k4x) / 6
        vy += (k1y + 2*k2y + 2*k3y + k4y) / 6
        vz += (k1z + 2*k2z + 2*k3z + k4z) / 6
        phi += (k1phi + 2*k2phi + 2*k3phi + k4phi) / 6
        
        x.append(x[-1] + h * vx)
        y.append(y[-1] + h * vy)
        z.append(z[-1] + h * vz)
        phi_list.append(phi)
    
    return x, y, z, phi_list

h = 0.01  # krok czasowy
steps = 1000  # liczba kroków czasowych

for phi0 in phi_values:
    x_rk4, y_rk4, z_rk4, phi_list = runge_kutta_4th_order(v0, theta, omega, np.radians(phi0), g, h, steps)
    plt.plot(x_rk4, y_rk4, label=f'fi_0 = {phi0} stopnii')

plt.xlabel('Odległość pałkarza od miotacza (m)')
plt.ylabel('y(m)')
plt.ylim(-2,2)
plt.xlim(0, 18.44)
plt.title('Trajektoria lotu knuckle balla w 2D dla różnych wartości początkowych położenia kątowego (Metoda Rungego-Kutty 4. rzędu)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

v0 = 29
theta = 4 
omega = 2
g = 9.81
v = 29

# Różne wartości początkowe dla phi
phi_values = [0, 18, 36, 54, 72]

def f(v):
    return 0.0039 + 0.0058 / (1 + np.exp((v - 35) / 5))

# Funkcja opisująca G(phi)
def G(phi):
    return 0.5 * (np.sin(4 * phi) - 0.25 * np.sin(8 * phi) + 0.08 * np.sin(12 * phi) - 0.025 * np.sin(16 * phi))

# Układ równań różniczkowych
def equations(v, phi):
    dvx_dt = -f(v) * v * np.cos(phi)
    dvy_dt = -f(v) * v * np.sin(phi) + g * G(phi)
    dvz_dt = -f(v) * v * np.sin(phi) - g
    dphi_dt = omega

    return dvx_dt, dvy_dt, dvz_dt, dphi_dt

# Metoda Rungego-Kutty czwartego rzędu (RK4)
def runge_kutta_4th_order(v0, theta, omega, phi0, g, h, steps):
    vx, vy, vz, phi = v0 * np.cos(np.radians(theta)), 0, v0 * np.sin(np.radians(theta)), phi0
    x, y, z, phi_list = [0], [0], [1.1], [phi]

    for _ in range(steps):
        k1x, k1y, k1z, k1phi = h * np.array(equations(v0, phi))
        k2x, k2y, k2z, k2phi = h * np.array(equations(v0, phi + k1phi / 2))
        k3x, k3y, k3z, k3phi = h * np.array(equations(v0, phi + k2phi / 2))
        k4x, k4y, k4z, k4phi = h * np.array(equations(v0, phi + k3phi))

        vx += (k1x + 2 * k2x + 2 * k3x + k4x) / 6
        vy += (k1y + 2 * k2y + 2 * k3y + k4y) / 6
        vz += (k1z + 2 * k2z + 2 * k3z + k4z) / 6
        phi += (k1phi + 2 * k2phi + 2 * k3phi + k4phi) / 6

        x.append(x[-1] + h * vx)
        y.append(y[-1] + h * vy)
        z.append(z[-1] + h * vz)
        phi_list.append(phi)

        if x[-1] >= 19.44:
            break

    return x, y, z, phi_list

h = 0.01  # krok czasowy
steps = 1000  # liczba kroków czasowych

# Symulacja i rysowanie wykresów w 3D
fig = plt.figure(figsize=(13,6.5))
ax = fig.add_subplot(111, projection='3d')

for phi0 in phi_values:
    x_rk4, y_rk4, z_rk4, phi_list = runge_kutta_4th_order(v0, theta, omega, np.radians(phi0), g, h, steps)
    ax.plot(x_rk4, y_rk4, z_rk4, label=f'fi_0 = {phi0} stopnii')

y = [-0.4, 0.4, 0.4, -0.4]
z = [0, 0, 1, 1]
x = [18.77, 18.77, 18.77, 18.77]

rect = Poly3DCollection([list(zip(x, y, z))], facecolors='red', edgecolors='r', alpha=0.7)
ax.add_collection3d(rect)

ax.set_xlabel('Odległość miotacza od pałkarza (m)')
ax.set_ylabel('Szerokość (m)')
ax.set_zlabel('Wysokość (m)')
ax.set_ylim(-1, 1)
ax.set_xlim(0, 18.44)
ax.set_zlim(0, 2)
ax.set_title('Trajektoria lotu knuckle balla w 2D dla różnych wartości początkowych położenia kątowego (Metoda Rungego-Kutty 4. rzędu)')
ax.legend(loc='upper left', bbox_to_anchor=(-0.5, 0.9))
plt.show()
```

Analizując wykres widzimy, jak bardzo nieregularna jest trajektoria *knuckle balla*. Rozumiejąc zasady baseballa jesteśmy w stanie stwierdzić, że miotacze wykorzystujący ten rodzaj rzutu są ogromnym zagrożeniem dla pałkarza, który nie jest w stanie przewidzieć toru, po którym piłka będzie poruszać się w jego stronę. W rezultacie, pałkarz ma mniejsze szanse na odbicie piłki.
